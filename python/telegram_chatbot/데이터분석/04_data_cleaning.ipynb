```python
import pandas as pd

df = pd.read_csv('./namsan.csv', encoding='EUC-KR', low_memory=False)
```

## 데이터삭제



```python
df.head(3)
```

```python
#슬라이싱으로 가능은 함
df.loc[:,'번호':'등록일자'].head()
```

```python
# 칼럼들
print(df.columns)
#마스킹생성
mask =df.columns !='Unnamed: 13'
print(mask)
#슬라이싱 

df.loc[:, mask]
```

```python
# 진짜로 칼럼 삭제한 df 리턴
df.drop(['부가기호', 'Unnamed: 13'], axis=1)

#리턴 말고 원본에서 삭제하기
df.drop(['부가기호','Unnamed: 13', '주제불류번호'],axis=1, inplace=True)
```

```python
df.head(3)
```

```python
# COL중에 하나라도  NAN(비어있음)이면 해당 COL 삭제
book_df = df.dropna(axis=1)
book_df.head(3)
# col의 모든 데이터가 NAN이면 해당 Col 삭제
book_df = df.dropna(axis=1, how='all')
book_df.head(3)
```
